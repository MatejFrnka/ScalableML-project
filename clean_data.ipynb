```python
import pandas as pd
import os
```

```python
BASEPATH = './raw_data/'
files = os.listdir(BASEPATH)

df_list = []
for file in files:
  if file == 'englandm.csv':
    df_list.append(pd.read_csv(BASEPATH + file, index_col=[0], low_memory=False))
    
    # Create standard: Home/Away team are referred to as Home and Away
    if 'HomeTeam' in df_list[-1].columns and 'AwayTeam' in df_list[-1].columns:
      df_list[-1]['Home'] = df_list[-1]['HomeTeam']
      df_list[-1]['Away'] = df_list[-1]['AwayTeam']
      df_list[-1].drop(['HomeTeam'], inplace=True, axis=1)
      df_list[-1].drop(['AwayTeam'], inplace=True, axis=1)

    # Create standard: Home/Away goals are referred to as FTHG and FTAG
    if 'HG' in df_list[-1].columns and 'AG' in df_list[-1].columns:
      df_list[-1]['FTHG'] = df_list[-1]['HG']
      df_list[-1]['FTAG'] = df_list[-1]['AG']
      df_list[-1].drop(['HG'], inplace=True, axis=1)
      df_list[-1].drop(['AG'], inplace=True, axis=1)

    # Create standard: Pinnacle odds are referred to as PH, PD and PA
    if 'PSH' in df_list[-1].columns and 'PSD' in df_list[-1].columns and 'PSA' in df_list[-1].columns:
      df_list[-1]['PH'] = df_list[-1]['PSH']
      df_list[-1]['PD'] = df_list[-1]['PSD']
      df_list[-1]['PA'] = df_list[-1]['PSA']
      df_list[-1].drop(['PSH'], inplace=True, axis=1)
      df_list[-1].drop(['PSD'], inplace=True, axis=1)
      df_list[-1].drop(['PSA'], inplace=True, axis=1)

    # Create standard: Home/Away goals are referred to as FTHG and FTAG
    if 'Div' in df_list[-1].columns:
      df_list[-1]['League'] = df_list[-1]['Div']
      df_list[-1].drop(['Div'], inplace=True, axis=1)

    # Create standard: Home/Away goals are referred to as FTHG and FTAG
    if 'Res' in df_list[-1].columns:
      df_list[-1]['FTR'] = df_list[-1]['Res']
      df_list[-1].drop(['Res'], inplace=True, axis=1)

df = pd.concat(df_list)
df
```

```python
for column in df:
  print(column, ':', df[column].isna().sum())
  if df[column].isna().sum() == 318728:
    df.drop([column], axis=1, inplace=True)

df
```

```python
df = df.sort_values('Date')
df
```

```python
premier_league_df = df.loc[df['League'] == 'E0']
df = df.loc[df['League'] != 'E0']
premier_league_df
```

```python
columns_to_keep = ['Date', 'Home', 'Away', 'PH', 'PD', 'PA', 'FTHG', 'FTAG', 'FTR', 'PSCH', 'PSCD', 'PSCA']

for column in df:
  if column not in columns_to_keep:
    df = df.drop(column, axis=1)
    premier_league_df = premier_league_df.drop(column, axis=1)

df
```

```python
for column in df:
  print(column, ':', df[column].isna().sum())
```

```python
premier_league_df = premier_league_df[premier_league_df['PSCH'].notna()]
premier_league_df[['PCH', 'PCD', 'PCA']] = premier_league_df[['PSCH', 'PSCD', 'PSCA']]
premier_league_df = premier_league_df.drop(['PSCH', 'PSCD', 'PSCA'], axis=1)

df = df[df['PSCH'].notna()]
df[['PCH', 'PCD', 'PCA']] = df[['PSCH', 'PSCD', 'PSCA']]
df = df.drop(['PSCH', 'PSCD', 'PSCA'], axis=1)
df
```

```python
premier_league_df = premier_league_df.reset_index(drop=True)

df = df.reset_index(drop=True)
df
```

```python
for column in df:
  print(column, ':', df[column].isna().sum())

premier_league_df = premier_league_df.dropna()
df = df.dropna()

for column in df:
  print(column, ':', df[column].isna().sum())
```

```python
premier_league_df.to_csv('./cleaned_data/premier_league.csv')
df.to_csv('./cleaned_data/england_cleaned.csv')
```
